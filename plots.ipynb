```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Loss Plots

```python
df_baselineloss = pd.read_csv(os.path.join('training_output', 'loss', 'baseline.csv')).drop('Wall time', axis=1).rename({'Value': 'Loss'}, axis = 1)
df_baselineloss['model'] = 'baseline'
df_config1loss = pd.read_csv(os.path.join('training_output', 'loss', 'config1.csv')).drop('Wall time', axis=1).rename({'Value': 'Loss'}, axis = 1)
df_config1loss['model'] = '1-Bit'

df_loss = pd.concat([df_baselineloss, df_config1loss])
df_loss
```

```python
sns.set_theme()
sns.lineplot(df_loss, x = 'Step', y = 'Loss', hue = 'model')
plt.title('Training Loss', weight='bold')
```

#### Validation Accuracy Plots

```python
df_baselineacc = pd.read_csv(os.path.join('training_output', 'accuracy', 'baseline.csv')).drop('Wall time', axis=1).rename({'Value': 'Accuracy'}, axis = 1)
df_baselineacc['model'] = 'baseline'

df_config1acc = pd.read_csv(os.path.join('training_output', 'accuracy', 'config1.csv')).drop('Wall time', axis=1).rename({'Value': 'Accuracy'}, axis = 1)
df_config1acc['model'] = '1-Bit'

df_acc = pd.concat([df_baselineacc, df_config1acc])
df_acc.head()
```

```python
sns.set_theme()
sns.lineplot(df_acc, x = 'Step', y = 'Accuracy', hue = 'model')
plt.title('Validation Accuracy', weight='bold')
```
